```python
import keras
from keras.datasets import mnist
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from keras.wrappers.scikit_learn import KerasClassifier
%matplotlib inline
```

```python
#MNISTを読み込み、trainとtestに分ける
(x_train, y_train), (x_test, y_test) = mnist.load_data()

#matplotlibで一部を表示
fig = plt.figure(figsize=(6, 6))
fig.subplots_adjust(left=0, right=0.8,  bottom=0, top=0.8, hspace=1, wspace=0.5)
for i in range(36):
    ax = fig.add_subplot(6, 6, i + 1, xticks=[], yticks=[])
    ax.imshow(x_train[i].reshape((28, 28)), cmap='gray')
```

```python
fig = plt.figure(figsize=(6, 6))
fig.subplots_adjust(left=0, right=0.8,  bottom=0, top=0.8, hspace=1, wspace=0.5)
for i in range(36):
    ax = fig.add_subplot(6, 6, i + 1, xticks=[], yticks=[])
    ax.imshow(x_train[i].reshape((28, 28)), cmap='gray')
    #正解ラベルを併記
    plt.title(y_train[i], color='red')
```

```python
img_rows = x_train[0].shape[1] #画像サイズを取得しておく
img_cols = x_train[0].shape[0]
num_classes = len(set(y_train)) #正解ラベルをset型にすることで何種類に分類しているのかを確認
```

```python
x_train = x_train.reshape(x_train.shape[0], img_rows, img_cols, 1)
x_test = x_test.reshape(x_test.shape[0], img_rows, img_cols, 1)
input_shape = x_train.shape[1:4]
```

```python
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
```

```python
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten
from keras.layers import Conv2D, MaxPooling2D

# 層を入れる順番は「畳み込み層→（ReLU）→プーリング層→畳み込み層→（ReLU）→プーリング層→全結合層→（ReLU）→全結合層→（ソフトマックス関数）」

model = Sequential()

#畳み込み層1
model.add(Conv2D(1, kernel_size=(3,3),
                 strides=1,
                 padding='same',
                 activation='relu',
                 use_bias = True,
                 input_shape=input_shape))

#プーリング層1
model.add(MaxPooling2D(pool_size=(2, 2),
                      strides=(1, 1),
                      padding='valid'))

#畳み込み層2
model.add(Conv2D(1, kernel_size=(3,3),
                 strides=1,
                 padding='same',
                 activation='relu',
                 use_bias = True,
                 input_shape=input_shape))

#プーリング層2
model.add(MaxPooling2D(pool_size=(2, 2),
                      strides=(1, 1),
                      padding='valid'))

#出力層
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(num_classes, activation='softmax'))


model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])
```

```python
model.fit(x_train, y_train,
          batch_size=128,
          epochs=3,
          verbose=1,
          validation_data=(x_test, y_test))
score = model.evaluate(x_test, y_test, verbose=1)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
